```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Create synthetic dataset
# -----------------------------
np.random.seed(50)

n = 20

# Raw values
age = np.random.normal(loc=40, scale=10, size=n)
distance = np.random.normal(loc=15, scale=5, size=n)

# Standardize (z-score)
age_z = (age - age.mean()) / age.std()
distance_z = (distance - distance.mean()) / distance.std()

# Performance classes
performance = np.random.choice(
    ["Exceeds", "Meets", "Does Not Meet"],
    size=n
)

df = pd.DataFrame({
    "age_z": age_z,
    "distance_z": distance_z,
    "performance": performance
})

# -----------------------------
# 2. Plot
# -----------------------------

color_map = {
    "Exceeds": "blue",
    "Meets": "green",
    "Does Not Meet": "orange"
}

fig, ax = plt.subplots(figsize=(15, 10))

for label in df["performance"].unique():
    subset = df[df["performance"] == label]
    ax.scatter(
        subset["age_z"],
        subset["distance_z"],
        label=label,
        color=color_map[label],
        s=60
    )

ax.set_xlabel("Standardized Age")
ax.set_ylabel("Standardized Distance from Home")
ax.set_title("Employee Performance by Age and Distance")
ax.legend(title="Performance Grade")
ax.grid(True)

plt.show()
```
